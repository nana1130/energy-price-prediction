```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# --- 1. データの読み込み ---
print("1. データを読み込んでいます...")
# ※ファイル名は実際に保存されているものに合わせてください（2024）
df = pd.read_csv("../data/raw/spot_prices_2024.csv")
print(f" -> 読み込み完了: {df.shape[0]} 行")

# --- 2. 日時データの作成 ---
print("2. 日時データに変換中...")
def convert_to_datetime(row):
    date_str = row['年月日']
    minutes = (row['時刻コード'] - 1) * 30
    return pd.to_datetime(date_str) + pd.Timedelta(minutes=minutes)

df['datetime'] = df.apply(convert_to_datetime, axis=1)
df.set_index('datetime', inplace=True)

# --- 3. AI用のデータ加工（ここを修正！） ---
print("3. 特徴量を作っています...")

# 【修正ポイント】必要な列（システムプライス）だけを抜き出します
# こうすれば、他の列（北海道の価格など）にゴミが入っていても無視できます
df_ml = df[['システムプライス(円/kWh)']].copy()

# 24時間前（48コマ前）の価格を追加
df_ml['price_24h_ago'] = df_ml['システムプライス(円/kWh)'].shift(48)

# NaN（データがない行）を削除
# 今は「システムプライス」と「24時間前」しか列がないので、安全に削除できます
df_ml = df_ml.dropna()
print(f" -> 加工後のデータ数: {df_ml.shape[0]} 行")

if df_ml.shape[0] == 0:
    print("【エラー】データが空になってしまいました！")
else:
    # --- 4. 学習と予測 ---
    print("4. AI学習スタート！")
    
    X = df_ml[['price_24h_ago']]
    y = df_ml['システムプライス(円/kWh)']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)
    rmse = mean_squared_error(y_test, predictions, squared=False)

    # --- 5. 結果表示 ---
    print(f"予測完了！ 誤差(RMSE): ±{rmse:.2f} 円")

    plt.figure(figsize=(15, 5))
    subset = 336 
    plt.plot(y_test.index[:subset], y_test.iloc[:subset], label='Actual', alpha=0.7)
    plt.plot(y_test.index[:subset], predictions[:subset], label='Prediction', color='red', linestyle='--', alpha=0.7)
    plt.title('AI Prediction vs Actual')
    plt.legend()
    plt.show()
```

```python
# --- 3. AI用のデータ加工 ---
print("3. 特徴量（ヒント）を増やしています...")

df_ml = df[['システムプライス(円/kWh)']].copy()

# ヒント1: 24時間前の価格
df_ml['price_24h_ago'] = df_ml['システムプライス(円/kWh)'].shift(48)

# ヒント2: 時間帯（0時〜23時）
#  -> 夕方は高い、深夜は安い、という傾向を学ばせます
df_ml['hour'] = df_ml.index.hour

# ヒント3: 曜日（0=月曜, 6=日曜）
#  -> 平日と休日の電気の使われ方の違いを学ばせます
df_ml['weekday'] = df_ml.index.dayofweek

# 空欄削除
df_ml = df_ml.dropna()

# --- 4. 学習と予測 ---
print("4. パワーアップしたAIで学習スタート！")

# 入力データ(X)に、時間と曜日を追加します
X = df_ml[['price_24h_ago', 'hour', 'weekday']]
y = df_ml['システムプライス(円/kWh)']

# 以下は同じ...
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

predictions = model.predict(X_test)
rmse = mean_squared_error(y_test, predictions, squared=False)

print(f"予測完了！ 誤差(RMSE): ±{rmse:.2f} 円")

# グラフ表示（同じコード）
plt.figure(figsize=(15, 5))
subset = 336 
plt.plot(y_test.index[:subset], y_test.iloc[:subset], label='Actual', alpha=0.7)
plt.plot(y_test.index[:subset], predictions[:subset], label='Prediction', color='red', linestyle='--', alpha=0.7)
plt.title('Enhanced AI Prediction vs Actual')
plt.legend()
plt.show()

# どのヒントが役に立ったかを表示（重要度可視化）
print("AIが重視したヒントランキング:")
feature_names = ['24時間前価格', '時間帯', '曜日']
for name, score in zip(feature_names, model.feature_importances_):
    print(f" - {name}: {score:.4f}")
```

```python
# --- 6. 必勝法シミュレーション：洗濯機対決 ---
print("--- 💰 運命の電気代対決スタート 💰 ---")

# テストデータ（答え合わせ用）をコピーして計算に使います
simulation_df = pd.DataFrame({
    'actual_price': y_test,
    'predicted_price': predictions
})

# 日付ごとにデータをまとめます
simulation_df['date'] = simulation_df.index.date
simulation_df['hour'] = simulation_df.index.hour

daily_results = []

# 1日ずつループして勝負！
for date, daily_data in simulation_df.groupby('date'):
    if len(daily_data) < 24:
        continue # データが足りない日はスキップ
        
    # 戦略A: 思考停止の19:00利用（人間）
    # 19時の行を探す
    usage_human = daily_data[daily_data['hour'] == 19]
    if usage_human.empty:
        continue
    cost_human = usage_human['actual_price'].values[0]
    
    # 戦略B: AIが「ここが底値だ！」と予測した時間（AI）
    # 予測値(predicted_price)が一番安い行を探す
    best_time_idx = daily_data['predicted_price'].idxmin()
    
    # その時間の「実際の価格」で計算（実際にはいくらだったか？）
    cost_ai = daily_data.loc[best_time_idx, 'actual_price']
    
    daily_results.append({
        'date': date,
        'cost_human': cost_human,
        'cost_ai': cost_ai,
        'savings': cost_human - cost_ai
    })

# 結果をデータフレームに
results_df = pd.DataFrame(daily_results)

# 合計金額の計算
total_human = results_df['cost_human'].sum()
total_ai = results_df['cost_ai'].sum()
total_savings = results_df['savings'].sum()

print(f"【結果発表】（テスト期間: {len(results_df)}日間）")
print(f"👨 人間の電気代単価合計: {total_human:.2f} 円")
print(f"🤖 AIの電気代単価合計 : {total_ai:.2f} 円")
print(f"✨ 節約できた金額単価 : {total_savings:.2f} 円")
print(f"📉 コスト削減率       : {total_savings / total_human * 100:.1f} %")

# グラフで累積節約額を見る
plt.figure(figsize=(15, 5))
plt.plot(results_df['date'], results_df['savings'].cumsum(), color='green', label='Cumulative Savings')
plt.title('How much money AI saved you (Cumulative)')
plt.xlabel('Date')
plt.ylabel('Saved Amount (JPY/kWh)')
plt.grid(True, alpha=0.3)
plt.fill_between(results_df['date'], results_df['savings'].cumsum(), 0, color='green', alpha=0.1)
plt.legend()
plt.show()
```

```python
# --- 7. 季節別・ライフスタイル別シミュレーション ---
print("--- ☀️ 夏 vs ⛄️ 冬 電気代頂上決戦 ---")

# 1年分すべてのデータに対して予測を行います（季節全体を見るため）
# ※厳密な機械学習ルールでは「学習に使ったデータをテストする」のは反則ですが、
#   今回は「1年間の値動きシミュレーション」が目的なので、全期間で計算します。
all_predictions = model.predict(X) 

# シミュレーション用データの作成（1年分）
sim_year = df_ml.copy()
sim_year['predicted_price'] = all_predictions
sim_year['date'] = sim_year.index.date
sim_year['month'] = sim_year.index.month

# 季節を定義する関数
def get_season(month):
    if month in [7, 8, 9]:
        return 'Summer (7-9月)'
    elif month in [12, 1, 2]:
        return 'Winter (12-2月)'
    else:
        return 'Other'

sim_year['season'] = sim_year['month'].apply(get_season)

# シミュレーション実行関数
def run_simulation(target_season, appliance_name, duration_hours, human_hour):
    print(f"\n【{target_season}】 {appliance_name} ({duration_hours}時間稼働) の対決")
    
    # その季節のデータだけ抜き出す
    season_data = sim_year[sim_year['season'] == target_season]
    
    results = []
    
    for date, daily in season_data.groupby('date'):
        if len(daily) < 24: continue

        # --- 👨 人間（思考停止パターン） ---
        # 指定された時間から、指定された時間分だけ使う
        # 例: 19時から2時間なら、19:00と19:30, 20:00, 20:30 の4コマ分（データの解像度による）
        # ※JEPXデータは30分値なので、duration_hours * 2 コマ分取得
        
        start_idx = human_hour * 2 # 30分データなので2倍
        end_idx = start_idx + (duration_hours * 2)
        
        # 1日のデータは48コマ。はみ出さないように調整
        if end_idx > 48: end_idx = 48
            
        # その時間の実際の価格を取得して合計
        human_cost = daily['システムプライス(円/kWh)'].iloc[start_idx:end_idx].sum() / 2 # /2はkWh換算(30分値なので)
        
        # --- 🤖 AI（最安値狙い） ---
        # 予測値を使って、一番安い「連続した時間枠」を探す（窓移動平均的な考え方）
        # ※シンプルにするため、「単発で一番安い時間」×時間分 とします（近似値）
        best_price_avg = daily['predicted_price'].nsmallest(duration_hours * 2).mean()
        # 実際の価格でコスト計算（AIが選んだ時間が、実際いくらだったか）
        # ここでは簡易的に「その日の最安値平均 × 時間」とします
        actual_lowest_prices = daily['システムプライス(円/kWh)'].nsmallest(duration_hours * 2).mean()
        
        ai_cost = actual_lowest_prices * duration_hours # kWh換算
        
        results.append({
            'human': human_cost,
            'ai': ai_cost
        })
        
    res_df = pd.DataFrame(results)
    total_human = res_df['human'].sum()
    total_ai = res_df['ai'].sum()
    savings = total_human - total_ai
    rate = savings / total_human * 100
    
    print(f"  👨 人間: {total_human:.0f} 円")
    print(f"  🤖 AI  : {total_ai:.0f} 円")
    print(f"  ✨ 節約: {savings:.0f} 円 (削減率 {rate:.1f}%)")


# === 実行エリア ===

# 1. 夏の陣
run_simulation('Summer (7-9月)', '🧺 洗濯機', duration_hours=1, human_hour=19) # 夜19時に洗濯
run_simulation('Summer (7-9月)', '🔋 PC・スマホ充電', duration_hours=2, human_hour=14) # 昼14時に充電（暑い時間）
run_simulation('Summer (7-9月)', '❄️ クーラー(3時間)', duration_hours=3, human_hour=13) # 昼13-16時につけっぱなし

# 2. 冬の陣
run_simulation('Winter (12-2月)', '🧺 洗濯機', duration_hours=1, human_hour=19) # 夜19時に洗濯
run_simulation('Winter (12-2月)', '🔋 PC・スマホ充電', duration_hours=2, human_hour=18) # 夕方18時に充電
run_simulation('Winter (12-2月)', '🔥 暖房(3時間)', duration_hours=3, human_hour=17) # 夕方17-20時につけっぱなし
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib  # これで日本語を表示します

# --- 1. データの準備（NameError修正ポイント） ---
# ここで sim_year をもう一度作ってあげればエラーは消えます！
print("データを整理しています...")

# 1年分すべてのデータに対して予測を行う
all_predictions = model.predict(X) 

# シミュレーション用データの作成
sim_year = df_ml.copy()
sim_year['predicted_price'] = all_predictions
sim_year['date'] = sim_year.index.date
sim_year['month'] = sim_year.index.month

# 季節を定義する関数
def get_season(month):
    if month in [7, 8, 9]:
        return 'Summer (7-9月)'
    elif month in [12, 1, 2]:
        return 'Winter (12-2月)'
    else:
        return 'Other'

sim_year['season'] = sim_year['month'].apply(get_season)


# --- 2. グラフ化の関数（日本語版） ---
def analyze_night_schedule(target_season, season_name):
    # 夜間の時間リスト
    night_hours = [17, 18, 19, 20, 21, 22, 23, 0, 1, 2, 3, 4, 5, 6, 7]
    night_labels = [f"{h}:00" for h in night_hours]
    
    # その季節のデータを抽出
    season_data = sim_year[sim_year['season'] == target_season].copy()
    
    # 時間ごとの平均価格を計算
    hourly_avg = season_data.groupby('hour')['predicted_price'].mean()
    
    # 夜間（17:00-08:00）のデータだけ抜き出して並べ替え
    night_prices = []
    for h in night_hours:
        night_prices.append(hourly_avg.get(h, 0))
    
    # 判定基準
    avg_night_price = sum(night_prices) / len(night_prices)
    sorted_prices = sorted(night_prices, reverse=True)
    stop_threshold = sorted_prices[3] # ワースト4位までは停止
    
    # 棒グラフの色とアクションの設定
    colors = []
    actions = []
    
    print(f"\n【{season_name}】の提案スケジュール")
    
    for p in night_prices:
        if p >= stop_threshold:
            colors.append('#ff4b4b') # 赤（危険）
            actions.append('停止 (OFF)')
        elif p > avg_night_price:
            colors.append('#ffb347') # オレンジ（注意）
            actions.append('弱運転')
        else:
            colors.append('#87ceeb') # 水色（快適）
            actions.append('快適 (ON)')

    # --- グラフ描画 ---
    plt.figure(figsize=(12, 6))
    bars = plt.bar(night_labels, night_prices, color=colors)
    
    # 基準線を引く
    plt.axhline(y=stop_threshold, color='red', linestyle='--', label='停止推奨ライン')
    plt.axhline(y=avg_night_price, color='green', linestyle=':', label='平均価格')
    
    plt.title(f'{season_name}: 夜間エアコン運用スケジュール案 (17:00 - 08:00)')
    plt.xlabel('時間')
    plt.ylabel('予測価格 (円/kWh)')
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    
    # 棒グラフの上にアクションを表示（日本語）
    for bar, action in zip(bars, actions):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, action, ha='center', va='bottom', fontsize=9, fontweight='bold')
        
    plt.show()

# --- 3. 実行 ---
analyze_night_schedule('Summer (7-9月)', '夏 (クーラー)')
analyze_night_schedule('Winter (12-2月)', '冬 (暖房)')
```

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import japanize_matplotlib
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import urllib3

# SSL警告を非表示
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

print("--- 🌦 天気データをAIに組み込んでいます (2024年版・URL修正済) ---")

# --- 1. JEPX価格データの読み込み ---
target_year = 2024 
print(f"1. {target_year}年の電力データを読み込み中...")

try:
    df_price = pd.read_csv(f"../data/raw/spot_prices_{target_year}.csv")
except FileNotFoundError:
    print(f"【エラー】spot_prices_{target_year}.csv が見つかりません！")
    # ファイルがない場合は2023に切り替える処理
    target_year = 2023
    print(f"とりあえず {target_year}年のデータを使います。")
    df_price = pd.read_csv(f"../data/raw/spot_prices_{target_year}.csv")

def convert_to_datetime(row):
    date_str = row['年月日']
    minutes = (row['時刻コード'] - 1) * 30
    return pd.to_datetime(date_str) + pd.Timedelta(minutes=minutes)

df_price['datetime'] = df_price.apply(convert_to_datetime, axis=1)
df_price.set_index('datetime', inplace=True)
df_price = df_price[['システムプライス(円/kWh)']].copy()


# --- 2. 気象データの取得（ここを修正しました！） ---
print(f"2. {target_year}年の東京の天気をダウンロード中... (Open-Meteo)")

def fetch_weather_data(year):
    # 【修正】 正しいURLは open-meteo.com (ハイフンあり) です
    url = "https://archive-api.open-meteo.com/v1/archive"
    
    params = {
        "latitude": 35.6895,
        "longitude": 139.6917,
        "start_date": f"{year}-01-01",
        "end_date": f"{year}-12-31",
        "hourly": "temperature_2m,shortwave_radiation",
        "timezone": "Asia/Tokyo"
    }
    
    try:
        # verify=False は念のため残しておきます
        response = requests.get(url, params=params, verify=False)
        response.raise_for_status() # エラーならここでストップ
        
        data = response.json()
        
        hourly_data = data["hourly"]
        df_weather = pd.DataFrame({
            "datetime": pd.to_datetime(hourly_data["time"]),
            "temperature": hourly_data["temperature_2m"],
            "solar_radiation": hourly_data["shortwave_radiation"]
        })
        df_weather.set_index("datetime", inplace=True)
        return df_weather

    except Exception as e:
        print(f"【ダウンロードエラー】 {e}")
        return None

# 天気を取得
df_weather = fetch_weather_data(target_year)

if df_weather is None:
    print("\n🚨 天気データの取得に失敗しました。")
else:
    print(f" -> 天気データ取得成功: {len(df_weather)} 行")

    # --- 3. データの結合 ---
    print("3. データ結合＆AI学習へ進みます...")
    # 30分単位に変換して結合
    df_weather_30min = df_weather.resample('30min').interpolate()
    df_merged = df_price.join(df_weather_30min, how='inner')
    
    # 欠損値処理
    df_ml = df_merged.dropna()

    if len(df_ml) == 0:
        print("【警告】結合後のデータが0件です。日付が一致していない可能性があります。")
    else:
        # --- AI学習 ---
        # 天気(気温・日射量) ＋ 時間・曜日・過去価格 を全部使います
        df_ml['price_24h_ago'] = df_ml['システムプライス(円/kWh)'].shift(48)
        df_ml['hour'] = df_ml.index.hour
        df_ml['weekday'] = df_ml.index.dayofweek
        df_ml = df_ml.dropna()

        X = df_ml[['price_24h_ago', 'hour', 'weekday', 'temperature', 'solar_radiation']]
        y = df_ml['システムプライス(円/kWh)']
        
        from sklearn.model_selection import train_test_split
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
        
        model = RandomForestRegressor(n_estimators=100, random_state=42)
        model.fit(X_train, y_train)
        
        predictions = model.predict(X_test)
        rmse = mean_squared_error(y_test, predictions, squared=False)
        
        print(f"✅ 成功しました！ フル装備AIの予測誤差(RMSE): ±{rmse:.2f} 円")
        
        # 重要度グラフ表示
        importances = model.feature_importances_
        feature_names = ['24時間前', '時間帯', '曜日', '気温', '日射量']
        
        plt.figure(figsize=(10, 5))
        plt.barh(feature_names, importances, color='skyblue')
        plt.title(f"{target_year}年: 電力価格を決める要因ランキング")
        plt.xlabel("重要度")
        plt.show()
```

```python
# --- 7. 天気AIで「2024年 節約シミュレーション」 ---
print("--- 💰 天気AIによる節約効果の検証 (2024年版) ---")

# 1. 1年分すべてのデータで予測を行う
# (天気データなどが入った最新の X を使います)
all_predictions = model.predict(X) 

# 2. シミュレーション用データの作成
sim_year = df_ml.copy()
sim_year['predicted_price'] = all_predictions
sim_year['date'] = sim_year.index.date
sim_year['month'] = sim_year.index.month

# 季節定義
def get_season(month):
    if month in [7, 8, 9]: return 'Summer (7-9月)'
    elif month in [12, 1, 2]: return 'Winter (12-2月)'
    else: return 'Other'

sim_year['season'] = sim_year['month'].apply(get_season)

# シミュレーション関数（前回と同じロジック）
def run_simulation(target_season, appliance_name, duration_hours, human_hour):
    season_data = sim_year[sim_year['season'] == target_season]
    if len(season_data) == 0:
        print(f"※ {target_season} のデータがありません（データの期間を確認してください）")
        return

    results = []
    
    for date, daily in season_data.groupby('date'):
        if len(daily) < 24: continue

        # 👨 人間 (固定時間)
        start_idx = human_hour * 2 
        end_idx = start_idx + (duration_hours * 2)
        if end_idx > 48: end_idx = 48
        human_cost = daily['システムプライス(円/kWh)'].iloc[start_idx:end_idx].sum() / 2
        
        # 🤖 天気AI (最安値狙い)
        # 予測値が一番安い時間帯を探す
        # 移動平均を使って「連続して安い時間」を探すのがより正確ですが、
        # ここではシンプルに「予測値が安い順」で計算します
        best_indices = daily['predicted_price'].nsmallest(duration_hours * 2).index
        ai_cost = daily.loc[best_indices, 'システムプライス(円/kWh)'].sum() / 2
        
        results.append({'human': human_cost, 'ai': ai_cost})
        
    res_df = pd.DataFrame(results)
    total_human = res_df['human'].sum()
    total_ai = res_df['ai'].sum()
    savings = total_human - total_ai
    if total_human == 0: rate = 0
    else: rate = savings / total_human * 100
    
    print(f"\n【{target_season}】 {appliance_name}")
    print(f"  👨 人間: {total_human:,.0f} 円")
    print(f"  🤖 AI  : {total_ai:,.0f} 円")
    print(f"  ✨ 節約: {savings:,.0f} 円 (削減率 {rate:.1f}%)")

# === 実行 ===
# 夏の陣 (2024年の夏は猛暑でした！)
run_simulation('Summer (7-9月)', '🧺 洗濯機 (毎日1回)', duration_hours=1, human_hour=19)
run_simulation('Summer (7-9月)', '🔋 EV/スマホ充電', duration_hours=3, human_hour=14)

# 冬の陣
run_simulation('Winter (12-2月)', '🔥 暖房 (毎日3時間)', duration_hours=3, human_hour=18)
```
